# Distribution plot

```python
iris.head()
```

```python
sns.residplot(x='sepal_length', y='sepal_width' , data=iris)
```

```python
sns.residplot(x='sepal_length', y='sepal_width' , data=iris, lowess=True)
```

```python
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris)
```

```python
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris , kind='kde')
```

```python
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris , kind='hex')
```

```python
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris , kind='hex' , color="#4CB391")
```

```python
#regression plot
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris , kind='reg' , color="#4CB391")
```

# Pair

```python
dots=sns.load_dataset('dots')
```

```python
dots.head()
```

```python
sns.pairplot(dots)  #it will pair up time,coherence and firing_rate to form 9 combinational graph
```

```python
sns.pairplot(dots , hue='choice')
```

```python
#sns.pairplot(dots , hue='gender')  #based on the male or female i.e gender we can also make graph if gender is present in it
```

```python
x=sns.pairplot(dots , hue='choice')
x.map_diag(sns.distplot)    #graph above diagonal are upper and lower one are lower 
```

```python
#note run all these 3 code in same cell
x=sns.pairplot(dots , hue='choice')
x.map_diag(sns.distplot) 
x.map_upper(sns.kdeplot)   #graph present above the diagonal are upper
```

```python
x=sns.pairplot(dots , hue='choice')
x.map_diag(sns.distplot) 
x.map_lower(sns.kdeplot)   #graph present below the diagonal are lower
```
