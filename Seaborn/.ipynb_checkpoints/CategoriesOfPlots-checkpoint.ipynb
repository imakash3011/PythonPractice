```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
#it is used to check whether all data it loaded successfully by taking some sample
iris.head()
```

#it is used to plot any one parameter to the graph
dist plot
rug plot
kde plot

# distplot

```python
sns.distplot(iris["sepal_width"])
```

```python
sns.rugplot(iris["sepal_length"])
```

# KDE Plot

```python
sns.kdeplot(iris["sepal_length"])
```

```python
sns.distplot(iris["sepal_width"],kde=False)
```

# Distribution plot

```python
iris.head()
```

```python
sns.residplot(x='sepal_length', y='sepal_width' , data=iris)
```

```python
sns.residplot(x='sepal_length', y='sepal_width' , data=iris, lowess=True)
```

```python
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris)
```

```python
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris , kind='kde')
```

```python
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris , kind='hex')
```

```python
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris , kind='hex' , color="#4CB391")
```

```python
#regression plot
sns.jointplot(x='sepal_length', y='sepal_width' , data=iris , kind='reg' , color="#4CB391")
```
