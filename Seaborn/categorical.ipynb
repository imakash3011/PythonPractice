```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pointplot(x='sepal_width', y='species' ,data=iris)
```

```python
sns.catplot(x='sepal_width', y='species' ,data=iris)    #note factorplot is renamed as catplot
```
