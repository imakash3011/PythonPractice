#subplot using object oriented 

```python
fig,axes = plt.subplots(1,2)    #1 row 2 column
```

```python
fig,axes=plt.subplots(3,4)
plt.tight_layout()   # to prevent overlapping
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout() 
```

```python
fig,axes=plt.subplots(1,2)
axes[0].plot(x,y,y,x)  #here it is working because we are using only 1D matrix but it will not work for 2,2 subplots
axes[1].plot(y,x)
plt.tight_layout()   
```

```python
fig,axes=plt.subplots(2,2)
axes[(0,1)].plot(x,y)    # for 2D matrix provid axes values inside the tuple
axes[(1,1)].plot(y,x)
plt.tight_layout()
```

Setting Labels with the object oriented technique

```python
fig=plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('x-label')
axes.plot(x,y)
```

```python
fig=plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.plot(x,y)
```

```python
fig=plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')
axes.set_title('the graph')
axes.plot(x,y)
```

```python
fig=plt.figure()
x1=fig.add_axes([0.1,0.1,0.8,0.8])
x1.set_xlabel('x-axis')
x1.set_ylabel('y-axis')
x1.set_title('the graph')
x1.plot(x,y)

x2=fig.add_axes([0.3,0.3,0.3,0.3])
x2.set_xlabel('x-axis')
x2.set_ylabel('y-axis')
x2.set_title('the graph')
x2.plot(y,x)
```

```python
fig , axes = plt.subplots(1,2, figsize=[5,7])
axes[0].plot(x,y)
axes[0].set_xlabel('x-axis')
axes[0].set_ylabel('y-axis')
axes[0].set_title('the graph')


axes[1].plot(y,x)
axes[1].set_xlabel('x-axis')
axes[1].set_ylabel('y-axis')
axes[1].set_title('the graph')
plt.tight_layout()
```
