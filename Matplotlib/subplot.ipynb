```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.linspace(0,10,20)
y=x*x
z=x+y
```

```python
plt.subplot(3,2,1)          #3-parameters :  (row  , column ,number of graphs)
plt.plot(x,x)

plt.subplot(3,2,2)
plt.plot(x,y)

plt.subplot(3,2,3)

plt.subplot(3,2,4)
plt.plot(x,y)

plt.subplot(3,2,5)

plt.subplot(3,2,6)
plt.plot(x,x)
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.tight_layout()   #to plot prevent the overlapping of graph
```
