```python
import numpy as np
```

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'key1':[1,2,3],'a':[1,2,3],'b':[5,6,7]})
df2=pd.DataFrame({'key1':[1,2,3],'c':[1,2,9],'d':[5,8,9]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='inner')
```

```python
pd.merge(df1,df2,how='inner',on="key1")
```

```python
pd.merge(df1,df2,how='outer',on="key1")
```

```python
df3=pd.DataFrame({'key1':[1,2,3],'a':[1,2,3],'b':[5,6,7]})
df4=pd.DataFrame({'key1':[1,4,5],'c':[1,2,9],'d':[5,8,9]})
```

```python
df3
```

```python
df4
```

```python
pd.merge(df3,df4,how='outer',on="key1")
```

```python
df5=pd.DataFrame({'key1':[1,2,3],'a':[1,2,3],'b':[5,6,7], 'key2':[5,2,3]})
df6=pd.DataFrame({'key1':[1,4,5],'c':[1,2,9],'d':[5,8,9], 'key2':[5,2,3]})
```

```python
pd.merge(df5,df6,how='outer',on="key1")
```

```python
pd.merge(df5,df6,how='outer',on=["key1","key2"])
```

```python
pd.merge(df5,df6,how='right',on=["key1","key2"])
```
